```python
import uuid
import time

config = {"configurable": {"thread_id": "abc123"}}
config['configurable']['thread_id'] = uuid.uuid4()
```

```python
async def get_answer(query):
    answers = []
    async for stream_mode in agent.astream({"user_query": query}, config, subgraphs=True, stream_mode=["messages","custom"]):
        mode_type, payload = stream_mode[1], stream_mode[2]

        if mode_type == "custom":
            answers.append(payload)
    
    return answers
        
```

**LLM as a judge**

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from typing import TypedDict, Annotated

class MetricScore(TypedDict):
    score: Annotated[float, ..., "Score: <0 - 1>"]

LLM = "openai/gpt-oss-20b"
BASE_URL = "https://api.groq.com/openai/v1"
GROQ_API_KEY = ""

judge = ChatOpenAI(
    model = LLM,
    api_key = GROQ_API_KEY,
    base_url = BASE_URL,
    streaming = True
).with_structured_output(MetricScore, method="json_schema", strict=True)
```

```python
correctness_prompt = ChatPromptTemplate.from_messages([
    ("system",
     "You are an automated evaluator.\n"
        "Your task is to compare a generated answer against an authoritative reference answer.\n"
        "Evaluate ONLY factual correctness and alignment with the reference.\n"
        "\nScoring rules:\n"
        "- 1.0 = Fully correct. Matches all key facts in the reference. No incorrect info.\n"
        "- 0.5 = Partially correct. Some key facts missing OR partially incorrect.\n"
        "- 0.0 = Incorrect. Conflicts with reference OR irrelevant.\n"
        "\nRequirements:\n"
        "- Ignore style, structure, and formatting.\n"
        "- Do not evaluate helpfulness, quality, grammar, or format.\n"
        "- Do not consider missing elaboration if core facts match.\n"
        "\nYou MUST return ONLY valid JSON in this exact schema:\n"
        "'score': <float from 0 to 1>"
    ),

    ("human",
     "USER QUESTION:\n{question}\n\n"
     "GENERATED ANSWER:\n{answer}\n\n"
     "REFERENCE ANSWER:\n{reference}\n\n"
     "Return JSON only."
    )
])

def correctness(question, generated_answer, reference_outputs) -> float:
    try:
        msgs = correctness_prompt.format_messages(
            question = question,
            answer = generated_answer,
            reference = reference_outputs
        )
        res: MetricScore = judge.invoke(msgs)
        return res["score"]
    except Exception as e:
        return -1 # < Error 
```

# Evaluation

```python
for idx in range(len(Evaluation_results)):
    print(Evaluation_results.iloc[idx]['question'])
    
    try:
        start = time.perf_counter()
        res = await get_answer(Evaluation_results.iloc[idx]['question'])
        end = time.perf_counter()
    except Exception as e:
        print(e)

    # Latency
    Evaluation_results.loc[idx , 'Latency'] = end - start

    # Filter_Out results
    output = {}
    try:
        for r in res:
            
            match(r['type']):
                case 'Guardrail':
                    output.update({"Guardrail": r['content']})
                    
                case 'ConverstationOutput':
                    output.update({"Proceed2Orchestration": r['content']['proceed2Orchestration']})
                    output.update({"Conversation": r['content']['conversation']})
                
                case 'Conversation':
                    output.update({"Proceed2Orchestration": r['content']['proceed2Orchestration']})
                    output.update({"Conversation": r['content']['conversation']})
                

                case 'FinalOutput':
                    output.update({"Final_Answer": r['content']['final_answer']})
        
        if output['Guardrail'] == 0 and output['Proceed2Orchestration'] == True:
            Evaluation_results.loc[idx , 'Factual_correctness'] = correctness(
                Evaluation_results.iloc[idx]['question'],
                output['Final_Answer'],
                Evaluation_results.iloc[idx]['expected_answer'])
            
            Evaluation_results.loc[idx , 'Final_Answer'] = output['Final_Answer']
            
        elif output['Guardrail'] == 0 and output['Proceed2Orchestration'] == False:
            Evaluation_results.loc[idx , 'Factual_correctness'] = correctness(
                Evaluation_results.iloc[idx]['question'],
                output['Conversation'],
                Evaluation_results.iloc[idx]['expected_answer'])
    except Exception as e:
        print(e)

    if output['Guardrail'] == 0:
        Evaluation_results.loc[idx , 'Conversation'] = output['Conversation']
        Evaluation_results.loc[idx , 'Proceed2Orchestration'] = output['Proceed2Orchestration']

    print(Evaluation_results.iloc[idx])

    if idx%10 == 0 and idx!=0 :
        Evaluation_results.to_csv(f"output_data{idx}.csv")

    # Delay to adjust for time-limit
    time.sleep(120)
    
```
